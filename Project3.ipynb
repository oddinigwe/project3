```python
#to enable visualizations 
%matplotlib inline

# First, import pandas, a useful data analysis tool especially when working with labeled data
import pandas as pd

# import seaborn, visualization library in python 
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)

# Next, we'll load the cereal dataset, which is in the specified directory below
chicagocrime = pd.read_csv("C:\\Users\\oddin\\Desktop\\Project3\\\\project3\\chicagocrime.csv")

# Next, display the first 20 rows and all columns of the iris dataframe, good way to see the colum headings for the dataset
chicagocrime.head(10)
```

```python
# to count the frequency of values for each primary crime type in the dataset
chicagocrime["Primary Type"].value_counts()
```

```python
# to count the frequency of values for Arrests and non-arrests for the various crime types in the dataset
chicagocrime["Arrest"].value_counts()
```

```python
#To create a pie chart showing primary crime types and years
fig,ax = plt.subplots(figsize=(12,12))
chicagocrime['Primary Type'].value_counts(sort=False).plot(kind='barh')
plt.title('Primary Crime Type and Community Area',fontsize=20)
plt.xlabel('Ward',fontsize=16)
plt.ylabel('Primary Type',fontsize=16)
```

```python
# To create a kdeplot which is a seaborn plot useful for looking at univariate relations 
# Creates and visualizes a kernel density estimate of the underlying feature

sns.FacetGrid(chicagocrime, hue="Primary Type", size=6) \
   .map(sns.kdeplot, "Year")\
   .add_legend()
```

```python
# Violin plot, unlike box plots, depict the density of the data
# Denser regions of the data are fatter, and sparser thiner in a violin plot
# further showing the distributions of the features 
sns.violinplot(x="Arrest", y="Year", data=chicagocrime, size=40)
```
